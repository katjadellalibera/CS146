```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

# Distributions
### 1. Gamma

```python
# Set up the probability distribution
alpha_1 = 2
beta_1 = 1
distribution_1 = stats.gamma(a=alpha_1, scale=1/beta_1)
# Set up the probability distribution with a second set of parameters
alpha_2 = 5
beta_2 = 2
distribution_2 = stats.gamma(a=alpha_2, scale=1/beta_2)

# Plot the two distributions
plt.figure(figsize=(8, 4))
x = np.linspace(0, 8)
plt.plot(x, distribution_1.pdf(x), color='blue', label='alpha=2\nbeta=1')
plt.plot(x, distribution_2.pdf(x), color='red', label='alpha=5\nbeta=2')
plt.xlabel('x')
plt.ylabel('probability density')
plt.title('Gamma Distributions')
plt.legend()
plt.show()
```

### 2. Normal

```python
# Set up the probability distribution
alpha_1 = 2
beta_1 = 1
distribution_1 = stats.norm(loc=alpha_1, scale=1/beta_1)
# Set up the probability distribution with a second set of parameters
alpha_2 = 5
beta_2 = 2
distribution_2 = stats.norm(loc=alpha_2, scale=1/beta_2)

# Plot the two distributions
plt.figure(figsize=(8, 4))
x = np.linspace(-1, 8)
plt.plot(x, distribution_1.pdf(x), color='blue', label='alpha=2\nbeta=1')
plt.plot(x, distribution_2.pdf(x), color='red', label='alpha=5\nbeta=2')
plt.xlabel('x')
plt.ylabel('probability density')
plt.title('Normal Distributions')
plt.legend()
plt.show()
```

### 3. Beta

```python
# Set up the probability distribution
alpha_1 = 2
beta_1 = 4
distribution_1 = stats.beta(a=alpha_1, b=beta_1)
# Set up the probability distribution with a second set of parameters
alpha_2 = 5
beta_2 = 10
distribution_2 = stats.beta(a=alpha_2, b=beta_2)

# Plot the two distributions
plt.figure(figsize=(8, 4))
x = np.linspace(0, 1)
plt.plot(x, distribution_1.pdf(x), color='blue', label='alpha=2\nbeta=4')
plt.plot(x, distribution_2.pdf(x), color='red', label='alpha=5\nbeta=10')
plt.xlabel('x')
plt.ylabel('probability density')
plt.title('Beta Distributions')
plt.legend()
plt.show()
```
