# Inferring mean and variance of normally distributed data

In this task you compute the posterior distribution over the mean $x$ and variance $\sigma^2$ of normally distributed data.

We also visualize samples from the prior and the posterior to better understand how our estimate over the parameters is updated with the data.

Below are function definitions for the normal-inverse-gamma distribution, as you saw them in your pre-class work.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Function definitions for the normal-inverse-gamma distribution.
# The parameters of the distribution (mu, lambda, alpha, beta) are as defined here:
# https://en.wikipedia.org/wiki/Normal-inverse-gamma_distribution

def norminvgamma_pdf(x, sigma2, mu, nu, alpha, beta):
    # The pdf of the normal-inverse-gamma distribution at x (mean) and sigma2 (variance).
    # Note that we use nu here instead of lambda, to match the notation used in the Wikipedia conjugate priors table.
    return (
        stats.norm.pdf(x, loc=mu, scale=np.sqrt(sigma2 / nu)) *
        stats.invgamma.pdf(sigma2, a=alpha, scale=beta))


def norminvgamma_rvs(mu, nu, alpha, beta, size=1):
    # Draw n samples from the normal-inverse-gamma distribution.
    # Returns a matrix where each row contains a (x, sigma2) sample.
    # Note that we use nu here instead of lambda, to match the notation used in the Wikipedia conjugate priors table.
    sigma2 = stats.invgamma.rvs(a=alpha, scale=beta, size=size)  # Sample sigma^2 from the inverse-gamma
    x = stats.norm.rvs(loc=mu, scale=np.sqrt(sigma2 / nu), size=size)  # Sample x from the normal
    return np.vstack((x, sigma2)).transpose()
```

## Task 2: Calculate posterior
1. Below are prior hyperparameters from your pre-class work. Run the code to see samples from the prior and an estimate of the prior 95% confidence interval over $x$ and $\sigma^2$.
2. Calculate the posterior distribution over $x$ and $\sigma^2$ using the prior hyperparameters below and the data set above.

```python
mu_0 = 0        # The prior mean is centered around 0.
nu_0 = 0.054    # The smaller nu_0 is, the more uncertain we are about the prior mean.
alpha_0 = 1.12  # alpha and beta govern the marginal prior over the variance.
beta_0 = 0.4


# Plot samples from the prior
num_samples = 10
samples = norminvgamma_rvs(mu_0, nu_0, alpha_0, beta_0, size=num_samples)

plt.figure(figsize=(12, 6))
plot_x = np.linspace(-15, 15, 500)
for i in range(num_samples):
    plot_y = stats.norm.pdf(plot_x, loc=samples[i, 0], scale=np.sqrt(samples[i, 1]))
    plt.plot(plot_x, plot_y)
plt.title('%i samples from a normal-inverse-gamma prior distribution' % num_samples)
plt.show()


# Estimate the 95% prior confidence intervals
samples = norminvgamma_rvs(mu_0, nu_0, alpha_0, beta_0, size=1000000)
print('95% prior confidence intervals:')
print('x: [%.3f, %.3f]' % (
    np.percentile(samples[:,0], 2.5),
    np.percentile(samples[:,0], 97.5)))
print(': [%.3f, %.3f]' % (
    np.percentile(samples[:,1], 2.5),
    np.percentile(samples[:,1], 97.5)))
```

```python
# Calculate the posterior distribution over mean x and variance 

# Calculate sufficient statistics
n = len(data)
data_sum = sum(data)
data_mean = data_sum / n

# Calculate posterior hyperparameters
mu_post = (nu_0 * mu_0 + data_sum) / (nu_0 + n)
nu_post = nu_0 + n
alpha_post = alpha_0 + n / 2
beta_post = beta_0 + sum((data - data_mean) ** 2) / 2 + (n * nu_0) / (n + nu_0) * (data_mean - mu_0) ** 2 / 2

print(mu_post, nu_post, alpha_post, beta_post)
```
